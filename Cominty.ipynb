# Setup ChromaDb as a Vector Database

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
import os

# Initialize embedding model
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Specify the directory for persistence
persist_directory = "./persistent_chroma_db"

# Ensure the directory exists
os.makedirs(persist_directory, exist_ok=True)

# Initialize Chroma with persistence
chroma_db = Chroma(
    persist_directory=persist_directory,  # Directory for persistent storage
    embedding_function=embeddings
)
```

```python
from langchain.document_loaders import PyPDFLoader

def extract_text_from_pdf(file_path):
    loader = PyPDFLoader(file_path)
    documents = loader.load()
    return documents
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Function to process and add PDFs to ChromaDB
def add_pdf_to_chroma_db(pdf_path, chroma_db):
    # Extract text from the PDF
    documents = extract_text_from_pdf(pdf_path)

    # Split text into chunks for better embedding
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50  # Overlapping helps maintain context
    )
    split_documents = text_splitter.split_documents(documents)

    # Add chunks to ChromaDB
    texts = [doc.page_content for doc in split_documents]
    metadatas = [{"source": pdf_path} for _ in split_documents]

    chroma_db.add_texts(texts=texts, metadatas=metadatas)
    chroma_db.persist()  # Save to persistent storage
    print(f"Added content from {pdf_path} to ChromaDB.")
```

# Define Query Routing Logic

```python
from langchain.schema import HumanMessage

def route_query_with_llm(query, llm):
    """Determine the appropriate resource (CSV or ChromaDB) for the query using the LLM."""
    routing_prompt = f"""
    You are a system designed to route user queries to the correct resource.

    - If the query involves structured tabular data (like rows, columns, tables, or sales figures), respond with "CSV".
    - If the query involves unstructured document data (like PDFs, reports, paragraphs, or pages), respond with "ChromaDB".
    - If the query doesn't match either category or is ambiguous, respond with "Unknown".

    Query: "{query}"
    Which resource should handle this query?
    """
    # Wrap the input query in a HumanMessage or use llm.predict
    response = llm.predict(messages=[HumanMessage(content=routing_prompt)])
    return response.strip()
```

```python
from typing import Annotated
from typing_extensions import TypedDict
from langgraph.graph import StateGraph,START,END
from langgraph.graph.message import add_messages
```

```python
class State(TypedDict):
  # Messages have the type "list". The `add_messages` function
    # in the annotation defines how this state key should be updated
    # (in this case, it appends messages to the list, rather than overwriting them)
  messages:Annotated[list,add_messages]

graph_builder=StateGraph(State)
```

# Adding an Agent to retrieve chunks from a vector Database

```python
from langchain.vectorstores import Chroma
from transformers import AutoTokenizer, AutoModel
from langchain_experimental.agents.agent_toolkits import create_vectorstore_agent
from langchain.tools import Tool
from langgraph.graph.message import add_messages
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq
import torch

# State structure
class State(TypedDict):
    messages: Annotated[list, add_messages]

# Initialize Groq LLM
llm = ChatGroq(
    temperature=0,
    groq_api_key='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',
    model_name="llama-3.1-70b-versatile"
)

# Lightweight Embedding Layer from HuggingFace
class HuggingFaceEmbeddings:
    def __init__(self, model_name="sentence-transformers/all-MiniLM-L6-v2"):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModel.from_pretrained(model_name)

    def embed_text(self, text: str) -> list:
        inputs = self.tokenizer(text, return_tensors="pt", padding=True, truncation=True)
        with torch.no_grad():
            outputs = self.model(**inputs)
        # Use the mean of token embeddings as the sentence embedding
        embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().tolist()
        return embeddings

# Initialize ChromaDB with HuggingFace embeddings
vector_db_path = "/content/vector_db"  # Path for persistent storage
embedding_model = HuggingFaceEmbeddings()
vectorstore = Chroma(
    persist_directory=vector_db_path,
    embedding_function=embedding_model.embed_text
)

# Path to your CSV file
csv_file_path = "/content/drive/MyDrive/supermarket_sales.csv"

# Create CSV Agent
csv_agent = create_csv_agent(
    llm,
    csv_file_path,
    verbose=True,
    allow_dangerous_code=True
)

# Define the CSV Query Tool
csv_tool = Tool(
    name="CSVQueryTool",
    func=lambda query: csv_agent.invoke(query),
    description="Tool for querying a CSV file."
)

# Create ChromaDB Agent
vectorstore_agent = create_vectorstore_agent(
    llm,
    vectorstore,
    verbose=True
)

# Define the Vectorstore Query Tool
vector_tool = Tool(
    name="VectorQueryTool",
    func=lambda query: vectorstore_agent.invoke(query),
    description="Tool for querying a vector database."
)

# Add tools to the LLM
llm_with_tools = llm.bind_tools(tools=[csv_tool, vector_tool])

# Initialize the graph builder
graph_builder = StateGraph(State)

# Add a chatbot node
def chatbot(state: State):
    return {"messages": [llm_with_tools.invoke(state["messages"])]}

# Add nodes and edges to the graph
graph_builder.add_node("chatbot", chatbot)
graph_builder.add_node("tools", ToolNode(tools=[csv_tool, vector_tool]))

# Add conditions for transitioning between nodes
def tools_condition(state: State) -> bool:
    return "use tools" in state["messages"]

graph_builder.add_conditional_edges("chatbot", tools_condition)
graph_builder.add_edge("tools", "chatbot")
graph_builder.add_edge(START, "chatbot")

# Compile the graph
graph = graph_builder.compile()

print("Graph successfully compiled with lightweight embeddings and ChromaDB integration.")
```
